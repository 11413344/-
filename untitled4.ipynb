```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Define the ticker symbol for TSMC (Taiwan Stock Exchange)
ticker_symbol = '2330.TW'

# Calculate the date for one year ago
end_date = datetime.now()
start_date = end_date - timedelta(days=365)

# Download historical data
tsmc_data = yf.download(ticker_symbol, start=start_date, end=end_date)

# Display the first few rows of the data
print("台積電最近一年的股價資料：")
display(tsmc_data.head())
```

```python
# Plot the closing price
plt.figure(figsize=(14, 7))
plt.plot(tsmc_data['Close'], label='TSMC Close Price')
plt.title(f'台積電 ({ticker_symbol}) 最近一年股價走勢圖')
plt.xlabel('日期')
plt.ylabel('股價 (新台幣)')
plt.grid(True)
plt.legend()
plt.show()
```
